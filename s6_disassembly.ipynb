## Storing and loading names

```python
def looping_sum(start, end, increment):
    result = 0
    for item in range(start, end, increment):
        result += item
    return result

dis.dis(looping_sum)
```

```python
dis.dis(
"""
result = 0
for item in range(start, end, increment):
    result += item
"""
)
```

```python
import time

start = time.perf_counter()
looping_sum(1, 10000000, 1)
end = time.perf_counter()

print(end - start)
```

```python
import time

start = time.perf_counter()
result = 0
for item in range(1, 10000000, 1):
    result += item
print(result)
end = time.perf_counter()

print(end - start)
```

```python
def loopy():
    x = []
    for i in range(0,10):
        x.append(i)
    return x

dis.dis(loopy)
```

```python
def compy():
    x = [i for i in range(0,10)]
    return x

dis.dis(compy)
```

## Is, by environment

```python
def notebook_is():
    two_hundred_fifty_seven = 257
    two_hundred_fifty_seven_again = 257
    return two_hundred_fifty_seven is two_hundred_fifty_seven_again
```

```python
notebook_is()
```

```python
dis.dis(notebook_is)
```

```python

two_hundred_fifty_seven = 257
two_hundred_fifty_seven_again = 257

two_hundred_fifty_seven is two_hundred_fifty_seven_again
```
